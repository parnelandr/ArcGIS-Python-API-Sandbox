# ArcGIS Rest Access via Requests Library

```python
import requests,json,getpass
```

```python
resp = requests.post("https://mpi.maps.arcgis.com/sharing/rest/generateToken?f=json", data={"username":username,"password":password,"referer":"https://mpi.maps.arcgis.com"})
token = json.dumps(resp.json()["token"]) 
print(json.dumps(resp.json()["token"]))
```

```python
serviceURL = "https://services2.arcgis.com/28amRQMPTiEvaF1p/arcgis/rest/services/service_c1a4112eafe74329acc50113f745f35f/FeatureServer/1"
queryPrefix = "/query?token=" 
querySuffix = "&where=1%3D1&returnIdsOnly=False&returnCountOnly=false&outFields=*&f=json"
getQuery = requests.get(serviceURL + queryPrefix + token + querySuffix)
print(getQuery.json()['features'])
```
